```r
library(tidyverse)
library(data.table)
```

```r
sds <- apply(subset, 2, sd)
```

# Thresholded standard deviation

```r
length(which(sds<0.001))
```

# 3 Standard Deviation v. Range
The first column is a control because (7.50921224921181, 12.5434922643017) is a subset of (8.259691, 12.046431).

```r
range <- range(subset[,1])
lb_ub <- c(mean(unlist(subset[,1])) - 3*sd(unlist(subset[,1])), mean(unlist(subset[,1])) + 3*sd(unlist(subset[,1])))
all(sapply(range, function(x) inrange(x, lb_ub[1], lb_ub[2]))==TRUE)
```

```r
controls <- apply(subset, 2, function(x) {
    range <- range(x)
    lb_ub <- c(mean(unlist(x)) - 3*sd(unlist(x)), mean(unlist(x)) + 3*sd(unlist(x)))
    all(sapply(range, function(x) inrange(x, lb_ub[1], lb_ub[2]))==TRUE)
})
```

```r
length(which(controls==TRUE))
```

```r
length(controls)
```
