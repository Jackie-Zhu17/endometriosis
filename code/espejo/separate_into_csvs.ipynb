# Separating Excel File into Convenient CSVs
The five sheets in the excel file were not uniformly formatted. I had to switch and remove certain rows to be the column names of the csv files.

```r
library(readxl)
```

```r
data_path <- "../../data/Data_Sheet_3_Weighted Gene Co-expression Network Analysis of Endometriosis and Identification of Functional Modules Associated With Its Main Hallmarks.xlsx"
out_path  <- "../../data/csv/"
```

```r
sheet_names <- excel_sheets(data_path)
csv_names <- paste0(sapply(sheet_names, function(x) paste0(strsplit(tolower(x), " ")[[1]], collapse="_")), ".csv")
csv_names
```

```r
k <- 5
excel_file <- read_excel(data_path, sheet=k)
```

```r
head(excel_file)
```

```r
names(excel_file) <- excel_file[1,]
```

```r
head(excel_file)
```

```r
write.csv(excel_file, paste0(out_path, csv_names[k]))
```
